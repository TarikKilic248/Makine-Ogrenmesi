```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
#rasgele veri noktalari icin
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples = 50, centers = 2, random_state=6)
```

```python
clf = svm.SVC(kernel="linear")
```

```python
clf.fit(X,y)
```

```python
#egitilmis sinifin karar sinirlearini gorsellestirme
plt.scatter(X[:,0],X[:,1], c=y, s=30, cmap=plt.cm.Paired)

#karar sinirlarini cizme
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

#modelin karar sinirlarini olusturma
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy,xx)

xy = np.vstack([XX.ravel(), YY.ravel()]).T

#karar fonk olusturduk
Z = clf.decision_function(xy).reshape(XX.shape)

#karar sinirlarini ve destek vektorlerini cizmek
ax.contour(XX,YY,Z, colors="k", levels=[-1,0,1], alpha=0.5, linestyles=["--","-","--"])

#destek vektorleri gorsellestirme adimi
ax.scatter(clf.support_vectors_[:,0], clf.support_vectors_[:,1], 
           s=100, linewidth=1, facecolor="none", edgecolor="k")


plt.show()
```
