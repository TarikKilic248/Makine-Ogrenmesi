```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
```

```python
#rastgele sayi ureticini baslatma
np.random.seed(0) 
age = np.random.randint(20,80,30)
blood_pressure = np.random.randint(80,180,30)
cholesterol = np.random.randint(120,350,30)
disease = np.random.randint(0,2,30) #0 saglikli, 1 hasta

data = pd.DataFrame({"Yas":age, "Kan Basinci": blood_pressure,
                     "Kolesterol": cholesterol, "Hasta": disease})

data
```

```python
X = data.drop("Hasta", axis=1)
y = data["Hasta"]

#estimator agacin sayisi
rf_classifier = RandomForestClassifier(n_estimators=5, random_state=42)
rf_classifier.fit(X,y)
```

```python
plt.figure(figsize=(20,10))
tree.plot_tree(rf_classifier.estimators_[0], filled=True, feature_names=list(X.columns),
              class_names=["Saglikli", "Hasta"], fontsize=10, rounded=True, precision=2,
              label="all", impurity=True, proportion=True)
```
